```
gharkbandy=["Aamir", "Kashif", "Car", "Chair", "Apple"]
for banday in gharkbandy:
  print(banday)
```

```
# check if items exists
home = ["chair", "sofa", "bench"]
if "chair" in home:
   print("Yes, 'chair' mojood hai")
```

```
# list length
home = ["chair", "sofa", "bench"]
print(len(home))
```

```
#add an item in list
home = ["chair", "sofa", "bench"]
home.append("car")
print(home)
```

```
#add an item in 2nd position at list
home = ["chair", "sofa", "bench", "car"]
home.insert(1,"table")
print(home)
```

```
#How to delete an item
home = ["chair", "sofa", "bench", "car"]
del home[1]
print(home)
```

```
home = ["chair", "table", "sofa", "bench", "car"]
home.remove("car")
print(home)
```

```
#clear the list
home = ["chair", "table", "sofa", "bench", "car"]
home.clear()
print(home)
```

```
#copy list
home = ["chair", "table", "sofa", "bench", "car"]
newhome = home.copy()
print(newhome)
```

```
l1 = ["a", "b", "c"]
l2 = ["x","y","z"]
l3 = l1 + l2
print(l3)
```

```
l1 = ["a", "b", "c"]
l2 = ["x","y","z"]
for x in l2:
  l1.append(x)
print(l1)
```
