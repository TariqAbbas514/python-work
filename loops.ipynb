# while loop

```
#while loop
mango = 1
while mango < 6:
  print(mango)
  mango += 1
```

```
mango = 1
while mango < 6:
  print(mango)
  if mango == 3:
    break
  mango += 1
```

```
mango = 12
while mango < 60:
  print(mango)
  if mango == 50:
    break
  mango += 5
```

```
i = 1
while i < 6:
 print(i)
 i += 1
else:
  print("i is not less than 6")
```

```
malta = 0
while malta < 6:
  malta += 1
  if malta == 3:
    continue
  print(malta)
```

# For loop

```
#for loop
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  print(x)
```

```
for i in "Pakistan":
  print(i)
```

```
for i in "Pakistan":
 print(i)
 if i=="s":
  break
```

```
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  if x == "banana":
    break
  print(x)
```

```
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  if x == "banana":
    continue
  print(x)
```

```
for i in range(10):
  print(i)
```

```
for i in range(2,6):
 print(i)
```

```
for i in range(2,30,3):
 print(i)
```

```
for i in range(8):
 print(i)
else:
  print("Finished")
```

```
t = ["red", "big", "tasty"]
fruits = ["apple", "banana", "cherry"]
for x in t:
  for y in fruits:
    print(x,y)
```
