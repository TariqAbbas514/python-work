# Functions

```
#Functions
#How are you
def acha(name):
  print(name + " Kiya hal han")
acha("Tariq")
```

```
def acha(name, last):
  print(name + " " + last)
acha("Tariq", "Abbas")
```

```
def my_function(child3, child2, child1):
  print("The youngest child is " + child3)

my_function(child1 = "Emil", child2 = "Tobias", child3 = "Linus")
```

```
def my_function(**kid):
  print("His last name is " + kid["lname"])

my_function(fname = "Tobais", lname = "Refsnes")  
```

```
#Functions Detailed
def bacha(*kids):
  print("how many kids you have " + kids[2])
bacha("Abid", "Kashif", "Adil")
```

```
def my_function(country = "Norway"):
  print("I am from " + country)
my_function("Sweden")
my_function()
```

```
def myfood(food):
  for x in food:
    print(x)
fruits = ["apple", "Banana", "Orange"]
myfood(fruits)
```

```
#Return values
def myfunction(x):
  return 5 * x
print(myfunction(5))
```
